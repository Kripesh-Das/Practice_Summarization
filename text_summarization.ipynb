### Prompt Template Text Summarization

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

generic_template = """
Write the summary of the following speech:
Speech:{speech}
Translate the entire summary to {language}
"""

prompt = PromptTemplate(
    input_variables=["speech","language"],
    template=generic_template
)
prompt
```

```python
complete_prompt = prompt.format(speech=speech,language="Hindi")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.run({"speech": speech, "language": "Hindi"})
summary
```

1. Overview of Text Summarization Techniques in LangChain
Techniques: Stuff Document Chain, MapReduce Summarization, Refine Chain Summarization.
Focus on the differences between these techniques and their suitable use cases.

2. Stuff Document Chain Summarization
Description: Combines entire document content and sends it as a single prompt to the LLM model.
Use Case: Best for smaller documents that fit within the LLM's context window.
Challenges:
Not suitable for large documents (e.g., PDFs with 100+ pages).
Limited by context size (e.g., 4096 tokens in GPT-3.5).
When documents are too large, direct summarization becomes impractical.

3. MapReduce Summarization Technique
Description: Breaks the document into smaller chunks, summarizes each chunk, and then combines the summaries into a final output.
Use Case: Ideal for larger documents or files that exceed the LLM's context size.
Process:
Step 1: Divide the document into chunks.
Step 2: Summarize each chunk using a prompt template and an LLM model.
Step 3: Combine individual summaries into a final summary.
Types:
Single Prompt Template: Same prompt for each chunk.
Multiple Prompt Template: Additional prompt applied to combined summaries for more specific information (e.g., extracting titles or quotes).

4. Comparison: Stuff vs. MapReduce
Stuff Document Chain:
Directly sends all data to the LLM.
Suitable if it fits within the context window.
MapReduce:
Divides content into smaller chunks.
Suitable when the document size exceeds the context window.
Reduces complexity through iterative summarization.

5. Practical Considerations
Context Size Limitation: Understanding the token limit of the LLM is crucial (e.g., 4096 tokens for GPT-3.5).
Chunking Strategy: Essential for large documents to ensure efficient processing.
Use of Multiple Prompts: Helpful for refining the final output with specific details.

### Stuff Document Chain Text Summarization

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
template = """  Write a concise summary of the following speech,
Speech : {text}
"""
prompt = PromptTemplate(input_variables=["text"],
                        template = template)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain = load_summarize_chain(llm,chain_type="stuff",prompt=prompt,verbose=True)
output = chain.run(docs)
output
```

### Map Reduce Text Summarization

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_documents=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
chunks_prompt="""
Please summarize the below speech:
Speech:`{text}'
Summary:
"""
map_prompt_template=PromptTemplate(input_variables=['text'],
                                    template=chunks_prompt)
```

```python
final_prompt='''
Provide the final summary of the entire speech with these important points.
Add a Motivation Title,Start the precise summary with an introduction and provide the summary in number 
points for the speech.
Speech:{text}

'''
final_prompt_template=PromptTemplate(input_variables=['text'],template=final_prompt)
final_prompt_template
```

```python
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output=summary_chain.run(final_documents)
output
```
